```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from scipy import stats
```

```python
df = pd.read_csv('NIS_2012-5_Core+Sev+Hosp_Use.csv')
df = df.dropna(subset=['DIED'])
X_train, X_test = train_test_split(df, test_size=0.3, random_state=0)
```

```python
catvars = ['FEMALE', 'IHCA', 'PMR', 'CM_CHRNLUNG', 'CM_DMCX', 'CM_DM', 'CM_HTN_C', 'CM_CHF', 'CM_METS', 'HTRAN', 'AFIB', 'CM_TUMOR', 'CS', 'CM_ARTH', 'STROKE' ,'CM_COAG', 'PCI', 'CM_VALVE', 'VFIB', 'PPPM', 'HBLOCK', 'CM_HYPOTHY', 'CM_PULMCIRC', 'CM_LYTES', 'CM_RENLFAIL', 'CM_ULCER', 'CM_PERIVASC', 'CM_LYMPH', 'CAROT', 'CM_ANEMDEF', 'VSD', 'CM_BLDLOSS', 'CM_PSYCH', 'STEMI', 'NSTEMI', 'CM_WGHTLOSS', 'AKI', 'HOSP_LOCTEACH', 'CABG', 'HOSP_REGION', 'HOSP_BEDSIZE', 'CM_LIVER', 'CM_PARA', 'CM_NEURO', 'CM_OBESE', 'HEMOD', 'CM_DRUG', 'CM_DEPRESS', 'PICD', 'PMI', 'KCAD', 'PCABG', 'PPCI', 'CM_ALCOHOL', 'CM_AIDS', 'SMOK', 'ZIPINC_QRTL', 'PAY1', 'FEMALE', 'RACE', 'YEAR']
contvars = ['AGE']
display(set(df.columns) - set(catvars) - set(contvars))  # unused variables
```

```python
X_train['training'] = 1
X_test['training'] = 0
X = pd.concat([X_train, X_test])
```

```python
for i in [X, X[X['training'] == 1], X[X['training'] == 0], X[X['DIED'] == 0], X[X['DIED'] == 1]]:
    print(len(i))
```

### Chi Square categorical

```python
for vari in catvars:
    print(f"{'#'*30} {vari} {'#'* 30}")
    testdf = X.copy().dropna(subset=[vari])
    vis = pd.DataFrame()
    for i, j in [('', {1: 'Overall'}), ('DIED', {1: 'Deceased', 0: 'Alive'}), ('training', {0: 'Testing Set', 1: 'Training Set'})]:
        conti = pd.crosstab(X[vari], X[i] if i else 1)
        if i:
            p = round(stats.chi2_contingency(pd.crosstab(testdf[i],testdf[vari]))[1], 3)
            vis[f'pvalue ({i})'] = p if p >= 0.001 else '<0.001'
        a = ((conti / conti.sum(axis=0)) * 100).round(2)
        a = conti.applymap("{:,.0f}".format).astype(str) + ' (' + a.astype(str) + ')'
        a = a.rename(columns=j)
        vis = pd.concat([vis, a], axis=1)
    vis = vis[['Overall', 'Training Set', 'Testing Set', 'pvalue (training)', 'Alive', 'Deceased', 'pvalue (DIED)']]
    display(vis)
```

### Independent Ttest continuous

```python
for vari in contvars:
    print(f"{'#'*30} {vari} {'#'* 30}")
    testdf = X.copy().dropna(subset=[vari])
    vis = {}
    for i, j in [('', {1: 'Overall'}), ('DIED', {1: 'Deceased', 0: 'Alive'}), ('training', {0: 'Testing Set', 1: 'Training Set'})]:
        if not i:
            vis[j[1]] = f"{round(testdf[vari].mean(), 2)} ± {round(testdf[vari].std(), 2)}"
            continue
        for k, v in j.items():
            vis[v] = f"{round(testdf[testdf[i] == k][vari].mean(), 2)} ± {round(testdf[testdf[i] == k][vari].std(), 2)}"
            ptemp = {f'pvalue ({i})': round(stats.ttest_ind(testdf[testdf[i]==0][vari], testdf[testdf[i]==1][vari], equal_var=False)[1], 2)}
        display(pd.DataFrame(ptemp, index=[1]))
    vis = pd.DataFrame(vis, index = [1])
    vis = vis[['Overall', 'Training Set', 'Testing Set', 'Alive', 'Deceased']]
    display(vis)
```
